# Сложность


### Константная сложность - O(1)

На графике отлично видно, что по оси абсцис(x) значения увеличиваются. Это наше n, входные значения.  
Но на оси ординат(y) все без изменений.  
Это говорит о том, что при увеличении входных значений(n), количество операций, которые необходимы для выполнения задачи, 
не изменяется. Это очень быстрые алгоритмы.


```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 10)
def y(n): return n*0 + 1

plt.plot(n, y(n))
# plt.show()
```

## Логарифмическая сложность - O(log n)
Это достаточно быстрый класс алгоритмов. Примером может служить бинарный поиск.  
Когда на каждом шаге объем обрабатываемых данных уменьшается.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 1000, 1, dtype=int)
y = lambda n: np.log(n)

plt.plot(n, y(n))
```

### Линейная сложность - O(n)
Смысл линейной сложности в том, что нам приходится проходить по всему набору входных данных.   
Это не самый быстрый класс алгоритмов, но и не самый медленный.

Например, линейный поиск в массиве. Чтобы найти нужное значение приходится перебирать один элемент за другим.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 10)
y = lambda n: n

plt.plot(n, y(n))
```

# Квадратичная сложность
Это весьма медленные алгоритмы.  
Например, сортировка вставками использует вложенный цикл. В результате мы имеем два вложенных цикла.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 10)
y = lambda n: n*n

plt.plot(n, y(n))
```

# Экспоненциальная сложность - O(n!)
Это очень медленные алгоритмы. В качестве примера можно привести задачу о коммивояжере.

```python
import matplotlib.pyplot as plt
import numpy as np


n = np.arange(1, 11, 1, dtype=int)
y = list(map(np.math.factorial, n))

plt.plot(n, y)
```
